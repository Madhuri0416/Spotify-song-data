```python
import pandas as pd

data = pd.read_csv(r'C:\Users\ammu0\Downloads\Spotify data.csv')
df = pd.DataFrame(data)
df
```

```python
# Number of files submitted for the month of March
df['added'] = pd.to_datetime(df['added'], format='mixed', dayfirst=True)
march_files = len(df[df['added'].dt.month == 3])
print(f"Number of files submitted for the month of March: {march_files}")
```

```python
# Maximum duration for each year 
dfsubset = df[['year released','dur']]
dfsubset.groupby('year released')['dur'].max()
```

```python
# Number of songs released for each artist for all years.
song_count= df.groupby('artist').size().reset_index(name='song')
song_count
```

```python
# Number of songs released for each artist in 2010.
artist_2010 = df[df['year released'] == 2010].groupby('artist').size()
artist_2010
```

```python
# Number of songs released for each top genre in each year released.
songs_genre_year = df.groupby(['top genre', 'year released']).size().unstack(fill_value=0)
songs_genre_year
```
